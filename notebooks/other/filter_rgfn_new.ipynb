```python
from rgfn import ReactionDataFactory

data_factory = ReactionDataFactory(
    reaction_path='../../data/rgfn_new/chemistry.xlsx'
)
reactions = data_factory.get_reactions()
```

```python
from collections import defaultdict

n_ary_dict = defaultdict(lambda: 0)
for reaction in reactions:
    n_ary_dict[len(reaction.left_side_patterns)] += 1
n_ary_dict
```

```python
with open('../../data/rgfn_new_filtered/templates.txt', 'w') as f:
    for reaction in reactions:
        if len(reaction.left_side_patterns) <= 2:
            f.write(reaction.reaction + '\n')
```

```python
fragments = data_factory.get_fragments()
with open('../../data/rgfn_new_filtered/fragments.txt', 'w') as f:
    for fragment in fragments:
        f.write(fragment.smiles + '\n')
```
