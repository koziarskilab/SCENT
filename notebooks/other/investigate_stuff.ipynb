```python
from rgfn import ReactionEnv, ReactionDataFactory

reaction_env = ReactionEnv(
    data_factory=ReactionDataFactory(reaction_path='../../data/synflow/templates.txt', fragment_path='../../data/synflow/fragment_64k.txt'),
    max_num_reactions=4
)
```

```python
from rgfn.gfns.reaction_gfn.api.reaction_api import ReactionActionC
from rgfn.gfns.reaction_gfn.api.data_structures import Reaction, Molecule, AnchoredReaction


# reaction = AnchoredReaction("[OH+0,O-:5]-[C:3](=[O:4])-[C$([CH]([CX4])),C$([CH2]):2] >> [OH+0,O-:5]-[C:3](=[O:4])-[C:2]([Br:6])", anchor_pattern_idx=0, idx=0)
# input_molecule = Molecule("Nc1n[nH]c2ncc(C3=CCC[C@H](C(=O)O)C3)cc12")
# output_molecule = Molecule("Nc1n[nH]c2ncc(C3=CCC[C@](Br)(C(=O)O)C3)cc12")
# reaction = Reaction("[OH+0,O-:5]-[C:3](=[O:4])-[C$([CH]([CX4])),C$([CH2]):2] >> [OH+0,O-:5]-[C:3](=[O:4])-[C:2]([Cl:6])", idx=0)
# input_molecule = Molecule("Nc1n[nH]c2ncc(C3=CCC[C@H](C(=O)O)C3)cc12")
# output_molecule = Molecule("Nc1n[nH]c2ncc(C3=CCC[C@](Cl)(C(=O)O)C3)cc12")

reaction = AnchoredReaction("[OH+0,O-:5]-[C:3](=[O:4])-[C$([CH]([CX4])),C$([CH2]):2] >> [OH+0,O-:5]-[C:3](=[O:4])-[C:2]([Cl:6])", idx=0, anchor_pattern_idx=0)
input_molecule = Molecule("O=C(O)c1c[nH]c2ncnc(N[C@H]3C=CC[C@H](C(=O)O)C3)c12")
output_molecule = Molecule("O=C(O)c1c[nH]c2ncnc(N[C@H]3C=CC[C@](Cl)(C(=O)O)C3)c12")
action_exp = ReactionActionC(input_reaction=reaction, input_molecule=input_molecule, output_molecule=output_molecule, input_fragments=())
input_molecule.valid, output_molecule.valid
```

```python
state = ReactionStateA(molecule=output_molecule, num_reactions=2)
reaction_env.parallel_action_space_cache = {}
action_space = reaction_env.get_backward_action_spaces([state])[0]
for action in action_space.possible_actions:
    if action.input_reaction == reaction:
        print(action == action_exp)
        print(action.output_molecule == output_molecule)
        print(action.input_molecule == input_molecule)
action_space.get_idx_of_action(action_exp)
```
