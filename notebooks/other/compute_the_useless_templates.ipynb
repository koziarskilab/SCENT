```python
from rgfn import ReactionDataFactory

data_factory = ReactionDataFactory(
    reaction_path='../../data/rgfn_new1/chemistry.xlsx'
)
```

```python
useless_templates = []
for reaction in data_factory.get_reactions():
    for pattern in reaction.left_side_patterns:
        has_match = False
        for fragment in data_factory.get_fragments():
            if fragment.rdkit_mol.HasSubstructMatch(pattern.rdkit_pattern):
                has_match = True
                break
        if not has_match:
            useless_templates.append(pattern)
            break
len(useless_templates), len(data_factory.get_reactions())
```
