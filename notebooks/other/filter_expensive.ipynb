```python
from pathlib import Path
import json

import pandas as pd
base_dir = Path('../../data/synflow')
fragments_path = base_dir / 'fragment_128k.txt'
cost_dict = json.load(open(base_dir / 'fragment_128k_to_real_cost.json'))
fragments_to_cost = {}
with open(fragments_path, 'r') as f:
    for line in f:
        fragment = line.strip()
        fragments_to_cost[fragment] = cost_dict[fragment]
len(fragments_to_cost)
```

```python
sorted_fragments = [x[0] for x in sorted(fragments_to_cost.items(), key=lambda x: x[1])]
```

```python
for k in [96, 112]:
    top_k = sorted_fragments[:k * 1000]
    print(len(top_k))
    fragment_path = base_dir / f'fragment_{k}k_cheapest.txt'
    cost_path = base_dir / f'fragment_{k}k_cheapest_to_real_cost.json'
    with open(fragment_path, 'w') as f:
        for fragment in top_k:
            f.write(fragment + '\n')
    
    new_cost_dict = {k: cost_dict[k] for k in top_k}
    with open(cost_path, 'w') as f:
        json.dump(new_cost_dict, f)
            
```
