```python
import json
import random

cost_dict = json.load(open('../../data/synflow/fragment_64k_to_real_cost.json'))
cost_items_list = list(cost_dict.items())
len(cost_dict)
```

```python
random.seed(0)
random.shuffle(cost_items_list)
for k in [1, 2, 4, 8, 16, 32]:
    new_dict = dict(cost_items_list[:k * 1000])
    with open(f'../../data/synflow/fragment_{k}k_to_real_cost.json', 'w') as f:
        json.dump(new_dict, f)
        
    with open(f'../../data/synflow/fragment_{k}k.txt', 'w') as f:
        for key in new_dict.keys():
            f.write(key + '\n')
    
```
