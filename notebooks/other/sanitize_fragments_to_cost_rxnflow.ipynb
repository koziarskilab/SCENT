```python
import json
from pathlib import Path

base_dir = Path('../../data/rgfn_new_filtered')
input_path = base_dir / 'fragment_to_real_cost.json'
output_path = base_dir / 'fragment_to_real_cost_rxnflow.json'
fragment_to_cost = json.load(open(input_path, 'r'))
```

```python
from rdkit import Chem


def sanitize(smiles):
    mol = Chem.MolFromSmiles(smiles, replacements={"[2H]": "[H]"})

    try:
        Chem.SanitizeMol(mol)
    except Exception:
        return None
    smiles = Chem.MolToSmiles(mol)

    return smiles
```

```python
fragment_to_cost_sanitized = {sanitize(k): v for k, v in fragment_to_cost.items()}
json.dump(fragment_to_cost_sanitized, open(output_path, 'w'))
```
