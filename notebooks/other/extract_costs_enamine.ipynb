```python
# open .sdf file
import requests
from bs4 import BeautifulSoup

# List of URLs to scrape
urls = [
    "https://www.enaminestore.com/catalog/EN300-26862755"
]

# Function to scrape cost from a single URL
def scrape_cost(url):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an error for bad HTTP responses
        
        soup = BeautifulSoup(response.content, "html.parser")
        
        print(response.content)
        
        # Replace 'cost-class' with the actual class or tag used for cost
        cost_element = soup.find(class_="p.c-text")
        if cost_element:
            return cost_element.text.strip()
        else:
            return "Cost not found"
    except requests.RequestException as e:
        return f"Error fetching URL {url}: {e}"

# Iterate through URLs and print costs
for url in urls:
    cost = scrape_cost(url)
    print(f"URL: {url} | Cost: {cost}")

```
