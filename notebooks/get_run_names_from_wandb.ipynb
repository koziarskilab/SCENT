```python
from collections import defaultdict
import pandas as pd 
import wandb
import pyperclip
api = wandb.Api()

filter_query = {
    "display_name": {"$regex": "^rgfn_seh_rgfn_new_filtered_old_code_grid_1_less_reactions"},
    "state": "finished",
}
group_by = ['ConstantCostModel.use_dataset']
filter_group_keys = [('configs/envs/dataset/synflow_56k_cheapest.gin',)][:0]
runs = api.runs("stereochemistry/RGFN", filters=filter_query)

grouped_runs = defaultdict(list)
for run in runs: 
    if run.summary.get("_runtime", 0) > 600:
        group_key = tuple(run.config.get(key, None) for key in group_by)
        if filter_group_keys and group_key not in filter_group_keys:
            continue
        grouped_runs[group_key].append(run)

string_list = []
for group_key, runs in grouped_runs.items():
    print(group_key)
    string = ",".join([f'"{run.name}"' for run in runs])
    print(string)
    pyperclip.copy(string)
    # name = f'rgfn_penalty_ablation_' + '_'.join(map(str, group_key))
    # string_list.append(f'("{name}", "helios", "RGFN"): [')
    # string = ",".join([f'"{run.name}"' for run in runs])
    # string_list.append(string)
    # string_list.append('],')
# end_str = ''.join(string_list)
# pyperclip.copy(end_str)
# end_str
```
